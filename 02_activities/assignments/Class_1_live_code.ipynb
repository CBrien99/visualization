```python
#load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

Making Data so we have something to plot

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x,y)
ax.set_title('Total growth over time lets make this a really long title', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

# Modifying our data points

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.scatter( x,
            y,
            marker='*',          # NEW ADDITIONS
            color = "indigo")    # NEW ADDITIONS

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,
        y,
        marker='*',         
        color='indigo',      
        linestyle='--',      
        linewidth=2)        

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# color being the new addition
ax.plot(x,y,marker='*', color = '#742d09', linestyle = '--', linewidth = 2)
fig.show()
```

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y,marker='*',              
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000') 
ax.grid()

ax.grid(axis = 'y')
```
